```python
# Program to simulate a traffic light

color = input("Enter the traffic light color (Red/Yellow/Green): ").strip().lower()

if color == "red":
    print("Stop.")
elif color == "yellow":
    print("Ready to move.")
elif color == "green":
    print("Go.")
else:
    print("Invalid color.")
```
